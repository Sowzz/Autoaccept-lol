## **Document 1 : Formule de conjugaison**
---
>**⚠ les longueurs indiquées sont des valeurs algébriques ⚠**
>
>Un objet $\overline {AB} $  situé en A du centre O d' une lentille mince forme une image en A' telle que:
>$ \frac {1}{\overline {OA'}} -\frac {1}{\overline {OA}}=\frac {1}{\overline {f'}} $ où F' est le foyer principal image de la lentille convergente.

## **Document 2 : Exemple d'utilisation de la méthode Monte-Carlo avec une seule mesure expérimentale**
---
La méthode Monte-Carlo (en bleu) consiste à simuler numériquement, un très grand nombre de fois, l'expérience qui a été réalisée durant le TP. Vous trouverez ci-dessous un script python permettant d’obtenir l’incertitude sur la valeur de $\overline{OA'}$ et son explication :

```python
import numpy as np  #importation bibliothèque standard de calculs
N = 1000     #nombre de tirages simulés
Di=[]     #liste des valeurs calculées de la distance (Di) séparant le centre de la lentille de l'image
val_min = float(input("Valeur minimale estimée en cm "))
val_max = float(input("Valeur maximale estimée en cm "))
for k in range(N) :     # répétition N fois de l'opération qui suit
  Dsim=np.random.uniform(val_max, val_min)     #la valeur de OA' est prise au hasard entre 2 valeurs extrêmes (val1 et val2) choisies par l’expérimentateur
  Di.append(Dsim)     #la valeur simulée est ajoutée à la liste de Di
Di_moy=sum(Di)/N     #calcul de la valeur moyenne de Di
uDi = np.std(Di,ddof=1)     #calcul de l’écart-type et donc de l’incertitude sur Di
print("Valeur moyenne de Di",Di_moy,"cm.")
print("Incertitude sur Di",uDi,"cm.")    #Ecriture des valeurs de Di et uDi
```

## **Document 3 : Exemple d’utilisation de la méthode Monte-Carlo avec une formule.**
---
On cherche la valeur moyenne et l’incertitude associée à D qui se calcule par la formule suivante : $D = \frac{3A+B}{C}$

On utilisera le script suivant :
```python
import numpy as np
N = 1000     
D=[]     

for k in range(N) :     
  A=np.random.uniform(val_min,val_max)  #Les valeurs de A, B et C sont prises au hasard entre les valeurs extrêmes choisies par l’expérimentateur
  B=np.random.uniform(val_min,val_max)
  C=np.random.uniform(val_min,val_max)
  D_int=((3*A)+B)/C   # calcul de D
  D.append(D_int)     # ajout à la list
Dmoy=sum(D)/N
uD=np.std(D,ddof=1)
print("Valeur moyenne de D",Dmoy,"cm.")
print("Incertitude sur D",uD,"cm.")   
```
